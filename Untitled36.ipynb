```python
scal_columns = data[["Age", "Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"]]
scal_columns
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
scal_columns = std_scaler.fit_transform(scal_columns)
scal_columns
```

```python
num_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


scal_attribs = ["Age", "Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"]
drop_attribs = ["id", "Unnamed: 0"]
object_attribs = x_train.select_dtypes(include=["object"]).columns.tolist()

# Create a ColumnTransformer with the custom transformer
preprocessing = ColumnTransformer([
        ("scale", num_pipeline, scal_attribs),
        ("todrop", 'drop', drop_attribs),
        ("encode",OneHotEncoder() , object_attribs)],
    remainder="passthrough")
```

```python
test = train_data.copy()

data_tran = preprocessing.fit_transform(test)

print("test_data.shape :",data_tran.shape)
```

```python
x_test = test_data.drop(["satisfaction"], axis=1)
y_test = test_data[["satisfaction"]]
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = []

def show(model):
    y_pred = model.predict(x_test)
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print("\naccuracy_score",accuracy_score(y_pred,y_test))
    accuracy.append(accuracy_score(y_pred,y_test))
```

```python
y_1D = y_train2["satisfaction"].values.ravel()

from sklearn.linear_model import LogisticRegression

log_model = make_pipeline(preprocessing,
                         LogisticRegression(random_state=42,max_iter=1000, n_jobs=-1, ))

log_model.fit(x_train,y_1D)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_model = make_pipeline(preprocessing,
                         RandomForestClassifier(n_estimators=100, random_state=42))

forest_model.fit(x_train,y_1D)
```

```python
show(forest_model)
```

```python
feature_importances = forest_model.named_steps['randomforestclassifier'].feature_importances_


for score, name in zip(feature_importances, x_train.columns):
    print(round(score, 2), name)
```

```python
from sklearn.neural_network import MLPClassifier

mlp_model = make_pipeline(preprocessing,
                         MLPClassifier(random_state=42))

mlp_model.fit(x_train, y_1D)
```

```python
show(mlp_model)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

extra_model = make_pipeline(preprocessing,
                         ExtraTreesClassifier(n_estimators=100, random_state=42))

extra_model.fit(x_train,y_1D)

show(extra_model)
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_clf = make_pipeline(preprocessing,
                         AdaBoostClassifier(n_estimators=150, learning_rate=1.1, random_state=42))
ada_clf.fit(x_train, y_1D)

show(ada_clf)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = make_pipeline(preprocessing,
                         GradientBoostingClassifier(max_depth=12, n_estimators=12, learning_rate=0.3))

gb_clf.fit(x_train, y_1D)
show(gb_clf)
```

```python
from sklearn.ensemble import VotingClassifier

voting_clf = VotingClassifier(
    estimators=[
        ("extra", extra_model),
        ("rf", forest_model),
        ("mlp", mlp_model),
        ("gb", gb_clf)
    ])

voting_clf.fit(x_train, y_1D)
```

```python
show(voting_clf)
```

```python
voting_clf_soft = VotingClassifier(
    estimators=[
        ("extra", extra_model),
        ("rf", forest_model),
        ("mlp", mlp_model),
        ("gb", gb_clf)
    ], voting="soft")

voting_clf_soft.fit(x_train, y_1D)
show(voting_clf_soft)
```

```python
from sklearn.ensemble import StackingClassifier

stacking_clf = StackingClassifier(
    estimators=[
        ("extra", extra_model),
        ("rf", forest_model),
        ("mlp", mlp_model),
        ("gb", gb_clf)
    ],
    final_estimator = RandomForestClassifier(random_state=42),
    cv=5
)

stacking_clf.fit(x_train, y_1D)
```

```python
show(stacking_clf)
```
